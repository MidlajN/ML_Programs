```python
# -----Missing Value Imputation-----

# n_hos_beds have missing values 
# machine learning algorithms does not support missing values
# solutions are :
#     * Remove row with missing data
#     * Impute missing data with mean/median
#         methods are :
#             * Impute missing with 0
#             * Impute with mean/median/mode
#                 + for numerical value - mean/median
#                 + for categorial value - mode
#             * Segment based imputation
```

```python
df.info()
```

```python
# by using this, the missing datas will be filled with mean
df.n_hos_beds = df.n_hos_beds.fillna(df.n_hos_beds.mean())
```
