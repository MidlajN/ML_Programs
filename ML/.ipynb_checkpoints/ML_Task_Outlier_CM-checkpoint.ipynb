```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('Movie_collection.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.boxplot('Twitter_hastags')
```

```python
# There is an outlier in the Twitter_hastags 
```

```python
upper_bound = np.percentile(data.Twitter_hastags,[99][0])
```

```python
upper_bound
```

```python
data.Twitter_hastags[(data.Twitter_hastags > 3 * upper_bound)] = 3 * upper_bound
```

```python
data.describe()
```

```python
data.boxplot('Avg_age_actors')
```

```python
np.percentile(data['Avg_age_actors'],[1])[0]
```

```python
lower_bound = np.percentile(data.Avg_age_actors,[1][0])
lower_bound
```

```python
# Avg_age_actors has outlier
```

```python
data.Avg_age_actors[(data.Avg_age_actors < 0.3 * lower_bound)] = 0.3 * lower_bound
```

```python
data.describe()
```

```python
#  Time_taken has missing values
```

```python
data.Time_taken.mean()
```

```python
data.Time_taken = data.Time_taken.fillna(data.Time_taken.mean())
```

```python
data.info()
```

```python
np.percentile(data.Time_taken,[1][0])
```

```python
#  Time_taken has outliers
```

```python
lower_bound_T = np.percentile(data.Time_taken,[1][0])
lower_bound_T
```

```python
data.Time_taken[(data.Time_taken < 0.3 * lower_bound_T)] = 0.3 * lower_bound_T
```

```python
data.describe()
```

```python
# ------ Classification Models ------

# 1. sklearn
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clfs_lr =  LogisticRegression()
```

```python
x = data[['Budget']]
y = data['Start_Tech_Oscar']
```

```python
clfs_lr.fit(x,y)
```

```python
clfs_lr.coef_
```

```python
clfs_lr.intercept_
```

```python
# -------------------------------------------------------------
```

```python
# 2. statsmodel
```

```python
import statsmodels.api as sn
```

```python
x_const = sn.add_constant(x)
```

```python
import statsmodels.discrete.discrete_model as sm
```

```python
logit = sm.Logit(y,x_const).fit()
```

```python
logit.summary()
```
